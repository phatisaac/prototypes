```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wakiso = gpd.read_file(r"C:\Users\HP\OneDrive\Desktop\QGIS\Organisation_units.geojson")
cases = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\QGIS\Wakiso District.csv")
```

```python
cases = cases.rename(columns={'organisationunitname':'name'})
```

```python
# Merge on subcounty name
wakiso_cases = wakiso.merge(cases, 
                            left_on="name",   # adjust to match GeoJSON property
                            right_on="name")    
```

```python
wakiso_cases
```

```python
fig, ax = plt.subplots(figsize = (10,8))
wakiso_cases.plot(column = 'Cases',
                  cmap = 'Reds',
                  legend = True,
                  edgecolor = 'blue',
                  ax = ax
                 )
plt.show()
                  
```

```python
import plotly.express as px

fig = px.choropleth(
    wakiso_cases,
    geojson=wakiso,
    locations="name",              # must match GeoJSON property
    featureidkey="properties.name",# adjust to your file schema
    color="Cases",
    hover_name="name",             # âœ… shows names on hover
    projection="mercator"
)

fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(width = 800, height = 700)
fig.show()
```
